### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
x_line = np.linspace(0, 10, 100)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (top-left)
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Chart')

# Bar chart (top-right)
axs[0, 1].bar(categories, bar_values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot (bottom-left)
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})',
                       (x_scatter[i], y_scatter[i]),
                       textcoords="offset points", xytext=(5,5), ha='left')

# Pie chart (bottom-right)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio for pie chart

plt.tight_layout()
plt.show()
```
